### Test the model

```python
import keras_ocr
import matplotlib.pyplot as plt

# download pre-trained text detector and text recognizer models from keras
model = keras_ocr.pipeline.Pipeline()

# define function for testing the model
def test_model(image_path: str)->None:
    """Test the Keras-OCR model using the image located at `image_path`."""
    
    # read the image
    image = [ keras_ocr.tools.read(image_path) ]
    
    # recognize text from the image
    result = model.recognize(image)[0]
    
    # display the image
    plt.imshow(image[0])
    
    # print all text recognized within the image
    for word, boundary in result:
        print(word)
    
    # annotate the recognized text in the image
    keras_ocr.tools.drawAnnotations(image[0], result)
    
```

#### Test 1

```python
test_model("../../data/image_1.png")
```

#### Test 2

```python
test_model("../../data/image_5.png")
```

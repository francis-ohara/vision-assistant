### Test the package

```python
import pyttsx3

engine = pyttsx3.init("nsss")

input_text = "Hello World. This is a text to speech program."
engine.say(input_text)
engine.runAndWait()

# check the current rate of speech
rate = engine.getProperty("rate")

# reduce the rate of speech to 150
engine.setProperty("rate", 180)

engine.say(input_text)
engine.runAndWait()
```
